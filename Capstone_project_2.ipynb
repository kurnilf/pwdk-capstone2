### Data Validation

```python
#Kolom dengan tipe kategori
kolom_kategori = data_cleaned.select_dtypes(exclude='number').columns # mengamil nama kolom yang memiliki tipe data selain numerik
kolom_kategori = kolom_kategori.drop(['ID','Dt_Customer']) #drop kolom id dan Dt_customer karena tipe datanya bukan kategorikal
kolom_kategori
```

Terdapat 2 kolom yang bertipe kategori yaitu `Education` dan `Marital_Status`. Kita akan memvalidasi data dua kolom tersebut.

#### Education

```python
#Menghitung dan cek kategori kolom Education
data_cleaned[kolom_kategori[0]].value_counts()
```

Berdasarkan riset tentang jenjang pendidikan di US yang telah saya lakukan. Jenjang pendidikan di US secara garis besar bisa dikategori menjadi empat yaitu:
1. Basic
2. Undergraduate
3. Master
4. PhD  

Oleh karena itu, data graduation di rename menjadi undergraduate, 2n Cycle direname menjadi Master. 

```python
#Rename isi data kolom education
education_rename = {'Graduation':'Undergraduate', '2n Cycle':'Master'}
data_cleaned['Education'] = data_cleaned[kolom_kategori[0]].replace(education_rename)
data_cleaned[kolom_kategori[0]].value_counts()
```

#### Marital_Status

```python
# menghitung dan cek kategori marital status
data_cleaned[kolom_kategori[1]].value_counts()
```

Status yang dapat diterima secara lega yaitu `Married`, `Single`, `Divorce`, dan `Widow`. Sehingga berdasarka requirement tersebut:
1. Status `Alone` akan diubah menjadi `Single`.
2. Status `Together` akan diubah menjadi `Single`. `Togehter` berarti sedang dalam hubungan namun belum menikah secara resmi.
3. Status `Absurd` dan `Yolo` akan diubah menjadi status yang paling sering muncul (mode)

```python
#Cek mode di kolom Marital_Status
data_cleaned[kolom_kategori[1]].mode()
```

```python
marital_status_rename = {'Together':'Single', 'Alone':'Single', 'Absurd':'Married', 'YOLO':'Married'}
data_cleaned[kolom_kategori[1]] = data_cleaned[kolom_kategori[1]].replace(marital_status_rename)
data_cleaned[kolom_kategori[1]].value_counts()
```

### Penambahan Kolom

#### Total Spending

```python
#Menambah kolom total spending
data_cleaned["TotalSpend"] = data_cleaned[["MntWines","MntFruits","MntMeatProducts","MntFishProducts", "MntSweetProducts", "MntGoldProds"]].sum(axis=1)
```

#### Umur

Pada dataset ini periode pengambilan data tidak dinyatakan secara explicit. Oleh karena itu perhitungan umur akan menggunakan tanggal terakhir pada kolom `Dt_Customer`

```python
#Menambah kolom 
data_cleaned["Age"] = data_cleaned['Dt_Customer'].max().year-data_cleaned["Year_Birth"]
data_cleaned["Age"].describe()
```

#### Tanggungan

Tanggungan adalah jumlah `Kidhome` dan `Teenhome`

```python
data_cleaned['Tanggungan'] = data_cleaned[['Kidhome','Teenhome']].sum(axis=1)
data_cleaned['Tanggungan'].describe()
```

#### Age Bin

```python
#Membuat pengelompokan umur
age_bin = [0, 18, 25, 33, 40, 47, 150]
label_bin = ['0-17', '18-24', '25-32', '33-39', '40-46', '47+']
data_cleaned['Age_bin'] = pd.cut(data_cleaned['Age'], bins=age_bin, labels=label_bin, right=False)
```

# Analisis dan visualisasi data

## 1. Efektivitas campaign

```python
campain_column = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']
data_campaign = data_cleaned[campain_column].sum().reset_index()
data_campaign.rename(columns={'index':'Kampanye',0:'Jumlah Campaign Berhasil'}, inplace=True)
data_campaign['Kampanye'] = data_campaign['Kampanye'].replace({
    'AcceptedCmp1':'Kampanye ke-1',
    'AcceptedCmp2':'Kampanye ke-2',
    'AcceptedCmp3':'Kampanye ke-3',
    'AcceptedCmp4':'Kampanye ke-4',
    'AcceptedCmp5':'Kampanye ke-5',
    'Response':'Kampanye Terakhir',})
data_campaign

```

```python
data_campaign['Tingkat berhasil (%)'] = (data_campaign['Jumlah Campaign Berhasil'] / len(data_cleaned) * 100).round(2)
data_campaign.sort_values(by='Tingkat berhasil (%)', ascending=False)
```

```python
#fig, ax = plt.subplots(figsize=(10,5))
plt.figure(figsize=(12,8))
ax = sns.barplot(data_campaign.sort_values(by='Tingkat berhasil (%)', ascending=False), x='Kampanye', y='Tingkat berhasil (%)', hue='Kampanye', legend=False, palette='pastel')
ax.set_title("Efektivitas Kampanye")
plt.plot()
```

Insight:
1. Kampanye dengan tingkat keberhasilan paling tinggi adalah kampanye terakhir dengan tingkat keberhasilan sebesar 15.05%
2. Kampanye ke-1, Kampanye ke-3, Kampanye ke-4, Kampanye ke-5 punya tingkat keberhasilan yang hampir sama.
3. Kampanye k-2 punya tingkat keberhasilan paling rendah.

## 2. Kampanye vs pendapatan

```python
#Unpivot data kammpanye
campain_column = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']
df1 = pd.melt(data_cleaned, id_vars='ID', value_vars=campain_column, var_name='Kampanye ke-')
df1 ['Kampanye ke-'] = df1['Kampanye ke-'].replace({
    'AcceptedCmp1':'Kampanye ke-1',
    'AcceptedCmp2':'Kampanye ke-2',
    'AcceptedCmp3':'Kampanye ke-3',
    'AcceptedCmp4':'Kampanye ke-4',
    'AcceptedCmp5':'Kampanye ke-5',
    'Response':'Kampanye Terakhir',})
df1
```

```python
#Ekstrak data pendapatan
pendapatan_data = data_cleaned[['ID','Income']]

#Ekstrak data kampanya yang sukses
filter_key = df1['value']==1
kampanye_sukses = df1[filter_key] 
```

```python
#Merge data pendapatan dengan data kampanye
pendapatan_kampanye = kampanye_sukses.merge(pendapatan_data, how='inner', left_on='ID', right_on='ID')
pendapatan_kampanye
```

```python
plt.figure(figsize=(12,8))
ax = sns.boxplot(data=pendapatan_kampanye, x='Kampanye ke-', y='Income', hue= 'Kampanye ke-', legend=False, palette='pastel')
ax.set_ylabel('Pendapatan ($)')
ax.set_title('Profil tingkat pendapatan yang menerima kampanye')
```

Insight:
1. Pelanggan yang menerima kampanye terakhir punya range pendaptan yang luas. Berkisatar antara $10.000 - $100.000.
2. Range pendapatan kampanye ke-3 dan kampanye terakhir hampir sama dan nilai median pendapatan kampanye ke-3 lebih rendah daripada kampanye terakhir.
